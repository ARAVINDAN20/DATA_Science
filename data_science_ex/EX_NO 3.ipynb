```python
Ex.No:3  Data Visualization


Reg:URK22AI1026


Aim
Write a python program to demonstrate the data visualization for data science application.

Description

Data visualization is the process of translating large data sets and metrics into charts,
graphs and other visuals. The resulting visual representation of data makes it easier to identify
and share real-time trends, outliers, and new insights about the information represented in the
data.

Matplotlib

Matplotlib is an easy-to-use, low-level data visualization library that is built on
NumPy arrays. It consists of various plots like scatter plot, line plot, histogram, etc.
Matplotlib provides a lot of flexibility.

import pandas as pd
import matplotlib. pyplot as plt
Dot Plots
These plots provide a visual representation of a function y = f(x) defined by a set of
(x,y) points. Dot plots just show the data points

Line Plot

Line Chart is used to represent a relationship between two data X and Y on a different axis. It
is plotted using the plot () function.
plt.plot(x, y, linewidth=0, marker=‘o’, color=&#39;lightblue&#39;)
         
Bar Plot and Pie Chart
         
A bar plot and Pie Chart is a graph that represents the category of data with rectangular
bars with lengths and heights that is proportional to the values which they represent. It can be
created using the bar () method. Both work by partitioning a geometric whole, be it a bar or a
circle, into areas proportional to the frequency of each group. Partitioning each bar into pieces
yields the stacked bar chart.
plt.bar(y, x, color=[‘red’,’blue’,’orange’])
plt.pie(x, labels = y,autopct=&#39;%1.2f%%&#39;)
        
Scatter Plot
        
Scatter plots are used to observe relationships between variables and uses dots to
represent the relationship between them. The scatter () method in the matplotlib library is used
to draw a scatter plot. It is also called as scatter graph, scatter chart, scattergram, or scatter
diagram. Scatter diagram uses dots It is capable of showing thousands of bivariate (two
dimensional) points in a clear understandable manner. It is with two variables mainly for
numerical data. The first variable is independent (x) and the second variable depends on the

first (y). The relationship between two variables is called their correlation.
plt. scatter (x, y, c=&#39;red&#39;)
              
Histograms
              
Histograms are an accurate representation of frequency distribution of
numerical data; it relates only one variable. Divide the entire range of values into
a series of intervals(horizontal) and then count or frequency (vertical) how many
values fall in to each interval. Theheight of each bar depicts the frequency of each
range or bin. Bins are consecutive, non- overlapping intervals of a variable. As the
adjacent bins leave no gaps, the rectangles of histogram touch each other to
indicate that the original value is continuous. The biggest issue in building a
histogram is deciding on the right number of bins to use. To understand the shape
of the distribution, we should use enough number of bins
plt.hist(x, color=&#39;red’, edgecolor=&#39;white’, bins=5)

Box and whisker plots
         
A box and whisker plot—also called a box plot—displays the five-number
summary ofset of data. The five-number summary is the minimum, first quartile
(25%), median (50%), third quartile (75%), and maximum. Provide the graphical
representation of describe () in dataframe. In a box plot, we draw a box from the
first quartile to the third quartile. A vertical line goes through the box at the
median. The whiskers go from each quartile to the minimum or maximum
Df [[&#39;Column name&#39;]]. boxplot ()
```

```python
# 2. Draw a comparative bar chart for “Graduate” and “Married” columns against
# Loan_status.
df = pd.read_csv('loan_prediction.csv')

grouped = df.groupby(['Loan_Status', 'Education', 'Married']).size().reset_index(name='Count')

pivot = grouped.pivot(index=['Loan_Status'], columns=['Education', 'Married'], values='Count')
pivot.plot(kind='bar')
# plt.ylim(top=15)

plt.title('Loan Status Count Comparison')
plt.xlabel('Loan Status')
plt.ylabel('Count')
plt.show()
```

```python
# 3. Draw a horizontal bar chart for “Property_area” and “Loan_status”
df = pd.read_csv('loan_prediction.csv')

grouped = df.groupby(['Property_Area', 'Loan_Status']).size().unstack()

# Plot the horizontal bar chart
grouped.plot(kind='barh', stacked=True)
plt.title('Loan Status by Property Area')
plt.xlabel('Count')
plt.ylabel('Property Area')
plt.show()
```

```python
# 6. Draw the dot plot between Gender and Loan_amount
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np

# Create a DataFrame from the given loan dataset
data = {
    'Gender': ['Male', 'Male', 'Male', 'Male', 'Female'],
    'LoanAmount': [np.nan, 128.0, 66.0, 120.0, 141.0]  # Replace np.nan with the actual value
}
df = pd.DataFrame(data)

# Create a dot plot
plt.figure(figsize=(8, 6))
plt.scatter(df['Gender'], df['LoanAmount'], c='blue', marker='o')
plt.title('Dot Plot of Loan Amount by Gender')
plt.xlabel('Gender')
plt.ylabel('Loan Amount')
plt.show()
```
