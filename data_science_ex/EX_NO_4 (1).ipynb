```python
#6.Calculate the IQR using quantile and remove the outliers in work_year column using IQR
#URK22AI1026

q75, q25 = np.percentile(df['work_year'], [75, 25])
iqr = q75 - q25

Q1 = df['work_year'].quantile(0.25)
Q3 = df['work_year'].quantile(0.75)
IQR = Q3 - Q1
l = Q1 - 1.5 * IQR
b = Q3 + 1.5 * IQR

s1 = df.loc[(df['work_year'] >= l) & (df['work_year'] <= b)]
s1.boxplot()
```

```python
# 13. Plot the heatmap using the correlation for employee table and titanic dataset.
import pandas as pd
import matplotlib.pyplot as plt

# Assuming you have read the CSV files into DataFrames
employee_data = pd.read_csv("employee.csv") # Replace with the actual file path
titanic_data = pd.read_csv("titanic.csv") # Replace with the actual file path

# Display the correlation heatmap for the employee table
employee_correlation = employee_data.corr()
plt.imshow(employee_correlation, cmap='coolwarm', interpolation='none')
plt.colorbar()
plt.xticks(range(len(employee_correlation.columns)), employee_correlation.columns, rotation='vertical')
plt.yticks(range(len(employee_correlation.columns)), employee_correlation.columns)
plt.title("Correlation Heatmap - Employee Data")
plt.show()
```

```python
# 13. Plot the heatmap using the correlation for employee table and titanic dataset.
# Display the correlation heatmap for the Titanic dataset
titanic_correlation = titanic_data.corr()
plt.imshow(titanic_correlation, cmap='coolwarm', interpolation='none')
plt.colorbar()
plt.xticks(range(len(titanic_correlation.columns)), titanic_correlation.columns, rotation='vertical')
plt.yticks(range(len(titanic_correlation.columns)), titanic_correlation.columns)
plt.title("Correlation Heatmap - Titanic Dataset")
plt.show()
```
