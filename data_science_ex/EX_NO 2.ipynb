```python
29. Drop the tuples which are not capable of receiving the loan.
```

```python
import pandas as pd

df = pd.read_csv('loan_prediction.csv')

df['Eligible'] = (df['Credit_History'] > 0) & (df['Property_Area'] != 'Rural')
df.drop(df[df['Eligible'] == False].index, inplace=True)
df.reset_index(drop=True, inplace=True)

print(df)
print(df.info())
```
